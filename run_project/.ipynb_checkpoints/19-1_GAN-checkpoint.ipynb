### 19장 세상에 없는 얼굴 GAN, 오토인코더

**```GAN(Generative Adversarial Networks)```** : 생성적 적대 신경망
- 가상의 이미지를 생성하는 알고리즘
- 알고리즘 내부에서 '적대적'인 경합 진행
    - **```생성자(Generator)```** : 가짜를 만들어내는 파트
    - **```판별자(Discriminator)```** : 진위를 가려내는 파트

````python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense, Reshape, Flatten, Dropout
from tensorflow.keras.layers import BatchNormalization, Activation, LeakyReLU, UpSampling2D, Conv2D
from tensorflow.keras.models import Sequential, Model

import numpy as np
import matplotlib.pyplot as plt
````

#### 1. 가짜 제조 공장, 생성자
- **```생성자(Generator)```** : 가상의 이미지를 만들어 내는 공장
    - 랜덤 픽셀 값으로 채워진 가짜 이미지로 시작해서   
        판별자의 판별 결과에 따라 지속적으로 업데이트
<br>
<br>

- **DCGAN(Deep Convolutional GAN)**
    - 생성자: 컨볼루션 신경망(CNNN)이용
        - 차이점(vs CNN)
            - optimizer를 사용하는 최적화 과정이나 컴파일하는 과정, pooling X
            - padding 과정 O $\because$ 입력 크기와 출력 크기를 맞추기 위해서
                - 판별자가 비교할 '진짜'와 똑같은 크기가 되어야 하기 때문
                - ```padding = 'same'```으로 설정
        - **배치 정규화(Batch Normalization)**
            - DCGAN 학습에 반드시 필요한 옵션
            - 입력 데이터의 평균이 0, 분산이 1이 되도록 재배치하는 것
            - 다음 층으로 입력될 값을 일정하게 재배치하는 역할
            - 층의 개수가 늘어나도 안정적인 학습 가능
            - ```BatchNormalization()``` 함수 제공
        - 활성화 함수: ```ReLU()```
            - 판별자로 넘겨 주기 전: ```tanh()```함수 → 출력되는 값을 -1~1 사이로 맞출 수 있음

````python
# 생성자 모델 만들기
generator = Sequential()
generator.add(Dense(128*7*7, input_dim=100, activation=LeakyReLU(0.2)))
generator.add(BatchNormalization())
generator.add(Reshape((7, 7, 128)))
generator.add(UpSampling2D())
generator.add(Conv2D(64, kernel_size=5, padding='same'))
generator.add(BatchNormalization())
generator.add(Activation(LeakyReLU(0.2)))
generator.add(UpSampling2D())
generator.add(Conv2D(1, kernel_size=5, padding='same', activation='tanh'))
````

#### 생성자 코드 분석   

- ```generator.add(Dense(128*7*7, input_dim=100, activation=LeakyReLU(0.2)))```
    - input_dim=100 : 100차원 크기의 랜덤 벡터를 준비해 집어넣으라는 뜻
    - 7 * 7: 이미지의 최초 크기
        - 28x28이 아닌 이유: UpSampling2D() 함수 사용하기 때문 
        - ```UpSampling2D()```: 이미지의 가로, 세로 크기를 2배씩 늘려줌   
            → 작은 크기의 이미지를 점점 늘려 가면서 컨볼루션 레이어를 지나가게 하는 것이 특징   
<br>
<br>

- ```generator.add(Reshape((7, 7, 128)))```
    - 컨볼루션 레이어가 받아들일 수 있는 형태로 바꾸어 주는 코드
    - 함수의 input_shape 부분에 들어갈 형태로 정해줌
<br>
<br>

- ```generator.add(UpSampling2D())```   
    ```generator.add(Conv2D(64, kernel_size=5, padding='same'))```   
   
    ```generator.add(UpSampling2D())```   
    ```generator.add(Conv2D(1, kernel_size=5, padding='same', activation='tanh'))```   
    - 두 배씩 upsampling을 한 후 컨볼루션 과정 처리
    - 커널 크기 : 5(5x5 크기 커널)
    - padding = 'same' : 모자라는 부분 자동으로 0   
<br>
<br>

- ```generator.add(Activation(LeakyReLU(0.2)))```   
    - ReLU() 함수를 쓸 경우 학습이 불안정
<br>
<br>

- ```generator.add(BatchNormalization())```   
    - 배치 정규화 진행(데이터의 배치를 정규 분포로 만듦)   
<br>
<br>

- ```generator.add(Conv2D(1, kernel_size=5, padding='same', activation='tanh'))```   
    - 한 번 더 컨볼루션 과정을 거친 후 판별자로 값을 넘길 준비 마침
    - 활성화함수: tanh()

#### 2. 진위를 가려내는 장치, 판별자
- 컨볼루션 신경망 구조를 그대로 가지고 옴
    - 컨볼루션 신경망: 무언가를 구별하는 데에 최적화된 알고리즘이기 때문
- 진짜인지 가짜인지 **판별만** 해 줄 뿐, 자기 자신이 **학습을 해서는 안 됨**
    - 판별자가 얻은 가중치는 생성자로 넘겨 주어 생성자가 업데이트된 이미지를 만들도록 해야 함
    - 학습 기능을 꺼주어야 함.

````python
# 판별자 모델 만들기
discriminator = Sequential()
discriminator.add(Conv2D(64, kernel_size=5, strides=2, input_shape=(28, 28, 1), padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Conv2D(128, kernel_size=5, strides=2, padding="same"))
discriminator.add(Activation(LeakyReLU(0.2)))
discriminator.add(Dropout(0.3))
discriminator.add(Flatten())
discriminator.add(Dense(1, activation='sigmoid'))
discriminator.compile(loss='binary_crossentropy', optimizer='adam')
discriminator.trainable = False
````

#### 판별자 코드 분석
- ```discriminator.add(Conv2D(64, kernel_size=5, strides=2, input_shape=(28, 28, 1), padding="same"))```
    - stride 사용 이유: 커널 윈도를 여러칸 움직이게 함으로써   
        가로, 세로 크기가 더 줄어들어 새로운 특징 추출 가능

#### 3. 적대적 신경망 실행하기

<br>

- 판별자 정확도가 0.5에 가까워질때, 학습 종료

````python
# 생성자와 판별자 모델을 연결시키는 GAN 모델 만들기
ginput = Input(shape=(100, ))

dis_output = discriminator(generator(ginput))
gan = Model(ginput, dis_output)
gan.compile(loss='binary_crossentropy', optimizer='adam')
gan.summary()
````

#### 코드 분석
- ```ginput = Input(shape=(100, ))```
    - 랜덤한 100개의 벡터를 Input() 함수에 집어넣어 생성자에 입력할 ginput을 만드는 과정
<br>
<br>

- ```dis_output = discriminator(generator(ginput))```
    - 판별자의 참과 거짓 판별값
    - generator()에 ginput 입력 → 28x28 이미지가 discriminator()의 입력 값으로 들어감
<br>
<br>

- ```gan = Model(ginput, dis_output)```
    - Model() 함수를 이용해 GAN 모델 생성
<br>
<br>

- ```gan.compile(loss='binary_crossentropy', optimizer='adam')```
    - gan 모델 컴파일

````python
# 신경망을 실행시키는 함수 만들기
def gan_train(epoch, batch_size, saving_interval):
    # batch_size : 한 번에 몇 개의 실제 이미지와 가상 이미지를 판별자에 넣을지 결정하는 변수
    # saving_interval: 몇 번마다 한 번씩 저장할지 정하는 변수
    
# MNIST 데이터 불러오기
    # 앞서 불러온 MNIST를 다시 이용, 테스트 과정은 필요 없고 이미지만 사용할 것이기 때문에 X_train만 호출
    (X_train, _), (_, _) = mnist.load_data()
    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
    # 127.5를 빼준 뒤 127.5로 나눠서 -1~1 사이의 값으로 바꿈
    X_train = (X_train - 127.5) / 127.5
    true = np.ones((batch_size, 1))    # 모두 참(1)이라는 레이블 값을 가진 배열, batch_size 길이만큼 만들어 d_loss_real에 사용
    fake = np.zeros((batch_size, 1))
    
    for i in range(epoch):
        # 실제 데이터를 판별자에 입력
        #randint(a, b, c): a 부터 b까지의 숫자 중 하나를 랜덤하게 선택해 가져오는 과정을 c번 반복
        idx = np.random.randint(0, X_train.shape[0], batch_size) # 실제 이미지를 랜덤하게 불러옴
        imgs = X_train[idx]
        # train_on_batch(x, y): 입력 값(x)과 레이블(y)을 받아서 딱 한 번 학습을 실시해 모델 업데이트
        d_loss_real = discriminator.train_on_batch(imgs, true)
        
        # 가상 이미지를 판별자에 입력
        noise = np.random.normal(0, 1, (batch_size, 100)) # 생성자에 집어넣을 가상 이미지를 만듦, normal -> 실수
        gen_imgs = generator.predict(noise) # noise 값이 생성자로 들어간 결괏값
        d_loss_fake = discriminator.train_on_batch(gen_imgs, fake) # gen_imgs가 '모두 가짜(0)'라는 레이블을 붙임
        
        # 판별자와 생성자의 오차 계산
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake) # real , fake 더하고 나눈 평균이 판별자의 오차
        g_loss = gan.train_on_batch(noise, true)
    
        print('epoch:%d' % i, ' d_loss:%.4f' % d_loss, ' g_loss:%.4f' % g_loss)
        
        # 중간 과정을 이미지로 저장하는 부분, 정해진 인터벌만큼 학습되면
        # 그 때 만든 이미지를 gan_images 폴더에 저장하라는 뜻
        if i % saving_interval == 0:
            # r, c = 5, 5
            noise = np.random.normal(0, 1, (25, 100))
            gen_imgs = generator.predict(noise)
            
            # Rescale images 0 - 1
            gen_imgs = 0.5 * gen_imgs + 0.5
            
            fig, axs = plt.subplots(5, 5)
            count = 0
            for j in range(5):
                for k in range(5): 
                    axs[j, k].imshow(gen_imgs[count, :, :, 0], cmap='gray')
                    axs[j, k].axis('off')
                    count += 1
                    fig.savefig("gan_images/gan_mnist_%d.png" % i)

# 4,000번 반복되고, 배치 크기는 32, 200번마다 결과가 저장됨
gan_train(4001, 32, 200)
````
